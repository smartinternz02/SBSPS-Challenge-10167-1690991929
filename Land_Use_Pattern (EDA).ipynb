### 1. Importing the essential Python Libraries

```python
# Importing the essentail libraries

import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import plotly.express as px
import plotnine
```

### 2. Loading the Dataset

```python
df = pd.read_csv('Land_Use_Pattern.csv')
df = df.iloc[:,1:]
df.head()

## All the Area related column have area in Hectares
# 1. States/UTs: Our Indian States and UTs
# 2. Category: Area / Percentage to geographical Area in Hectares
# 3. Total geographical area
# 4. Forests: Area under forest out of Total geographical area
# 5. Not available for cultivation: Area not available for cultivation out of Total geographical area
# 6. Permanent pastures and other grazing lands out of Total geographical area
# 7. Land under miscellaneous tree crops & groves out of Total geographical area
# 8. Culturable wasteland: Cultivable wasteland that is suitable for agriculture but is not exploited for one reason or another. 
#    This property was abandoned owing to a variety of factors, including a lack of clean water, saline soils or alkaline, 
#    erosion, waterlogging, an unfavorable physiographic situation, human neglect, or soil inadequacies caused by poor agricultural techniques.
# 9. Fallow lands other than current fallows:  This includes all land which was taken up for cultivation 
#    but is temporarily out of cultivation for a period of not less than one year and not more than five years.
# 10. Current Fallows: Current fallow land is land that is left uncultivated for one or less than one agricultural year.
# 11. Net Area Sown out of Total geographical area
```

### 5. Data Manipulation

```python
df['States/UTs'].value_counts()
```

```python
## Changing some values of States and UTs so that they can be recognised by IBM Cognos Analytics
df['States/UTs'] = df['States/UTs'].apply(lambda x:"Dadra and Nagar Haveli" if x == "Dadra & Nagar Haveli" else x)
```

### 6. Univariate Analysis

```python
df.columns
```

```python
df['Category'].value_counts()
```

```python
df_ = df[['States/UTs', 'Category', 'Total geographical area']]
df_ = df_[df_['Category'] == "Area"]
df_ = df_.sort_values(by = 'Total geographical area',ascending = False)

plt.pie(df_['Total geographical area'],labels = df_['States/UTs'],autopct = "%0.0f%%")
# Draw a white circle in the center to create the hole
center_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

plt.title("Overview of Total Geographical Area of States in India")
plt.show()
```

```python
df_ = df[['States/UTs', 'Category', 'Net area sown']]
df_ = df_[df_['Category']== "Area"]
df_ = df_.sort_values(by = 'Net area sown',ascending = False)[0:15]

plt.pie(df_['Net area sown'],labels = df_['States/UTs'],autopct = "%0.0f%%")
# Draw a white circle in the center to create the hole
center_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)


plt.title("Top 15 Net area sown in Hectares in States of India")
plt.show()
```

### 7. Bivariate Analysis

```python
df_ = df[['States/UTs', 'Category', 'Culturable wasteland']]
df_ = df_[df_['Category']== "Area"]
df_ = df_.sort_values(by = 'Culturable wasteland',ascending = False)[0:15]

sns.set_theme(font_scale = 0.8)
plt.figure(figsize = (14,7))
sns.barplot(x =df_['States/UTs'],y = df_['Culturable wasteland'],color = "green")
plt.title("Top 15 Culturable Wasteland Areas in States of India")
plt.show()
```

```python
df_ = df[['States/UTs', 'Category', 'Current fallows']]
df_ = df_[df_['Category']== "Area"]
df_ = df_.sort_values(by = 'Current fallows',ascending = False)[0:10]

sns.set_theme(font_scale = 0.8)
plt.figure(figsize = (14,7))
sns.barplot(x =df_['States/UTs'],y = df_['Current fallows'])
plt.title("Top 10 Current fallows Areas in States of India")
plt.show()
```

### 8. Multivariate Analysis

```python
df_ = df[["States/UTs",'Fallow lands other than current fallows', 'Category', 'Current fallows']]
df_ = df_[df_['Category'] == "Area"]
df_ = df_.dropna()

sns.set_theme(font_scale = 0.8)
plt.figure(figsize=(10, 6))

ax = sns.scatterplot(x = df_['Fallow lands other than current fallows'],y = df_['Current fallows'],hue =df_["States/UTs"] )
sns.move_legend(ax, "upper left",bbox_to_anchor = (1,1))
plt.title('Scatterplot of Condition on Fallows in States of India')
plt.show()
```
